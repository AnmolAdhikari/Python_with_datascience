### Linear regression using `sklearn`.
!pip install scikit-learn

````python
import numpy as np
from sklearn.linear_model import LinearRegression
````

````python
x_val = np.array([4,3,3,3,6,4,5,4,5,1,2,2,3]).reshape((-1, 1)) # Two dim
y_val = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
````

````python
model = LinearRegression().fit(x_val,y_val)
````

````python
model
````

**`r square` value**
- Represents the goodness-of-fit of a linear regression model.
- Indicates the proportion of the variance in the dependent variable (`y`) that can be explained by the independent variables(`x`).
- The value of R ranges from 0 to 1, where 0 indicates no linear relationship, and 1 indicates a perfect fit.

````python
model.score(x_val,y_val)
````

**y-intercept**

````python
intercept = model.intercept_
intercept
````

**slope**

````python
slope = model.coef_
slope
````

**Linear function and plot**

````python
def myfunc(x):
    return slope * x + intercept # y = mx +c
````

````python
myfunc(5)
````

````python
mymodel = list(map(myfunc, x_val))
mymodel
````

````python
import matplotlib.pyplot as plt
plt.scatter(x_val, y_val) ## scatter plot
plt.plot(x_val, mymodel, 'red') ## line plot
plt.show()
````

````python
myfunc(13)
````

### Pandas Exercise
- Create the `LinearRegression` model for `Year` vs `USA` Petrol Price.
- Find `c`, `m`, and `r square`.
- Plot the scatter as well as regression line graph.
- Predict the price in 2022 using linear model.

````python
import numpy as np
import pandas as pd
petrol=pd.read_csv('petrol_prices.csv')
````

````python
petrol
````

````python
a = np.array(petrol['Year']).reshape((-1, 1))
b = petrol['USA']
model = LinearRegression().fit(a,b)
````

````python
model
````

Computing `r square`

````python
model.score(a,b)
````

Computing `intercept`

````python
c = model.intercept_
c
````

Computing `slope`

````python
m = model.coef_
m
````

````python
a
````

````python
def my_model(x):
    return m*x+c

list(map(my_model,petrol['Year']))
````

````python
plt.title('USA Petrol Price per Gallon')
plt.scatter(petrol['Year'], petrol['USA'])
plt.plot(petrol['Year'],list(map(my_model,petrol['Year'])), 'red')
````

````python
my_model(2022)
````
