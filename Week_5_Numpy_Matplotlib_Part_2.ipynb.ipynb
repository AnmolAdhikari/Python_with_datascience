## Plotting libraries

Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]

y = [1,4,9,16,25,36,49]

plt.title("Our first graph")
plt.xlabel("This is x")
plt.ylabel("This is y")

plt.plot(x,y)
```

```python
x = [1,2,3,4,5,6,7]

y = [1,4,9,16,25,36,49]

plt.title("Our first graph", {'fontsize':25, 'color': 'blue'})

plt.xlabel("This is x", {'fontsize':15, 'color': 'orange'})

plt.ylabel("This is y")

plt.plot(x,y,color='green')
```

### Multiple lines?

```python
x = [0,1,2,3,4,5]

y = [5,10,15,13,11,10]

plt.title("Our first graph", fontdict = {'fontname':'DejaVu Sans', 'fontsize':40, 'color': '#d08d88'}) #fontdict = {'fontname':'DejaVu Sans', 'fontsize':20}

plt.xlabel("This is x", fontdict = {'fontsize': 30})

plt.ylabel("This is y")

plt.plot(x,y,color='red') # color, linewidth

a = [1,3,6,8]
b = [1,6,10,11]
plt.plot(a,b)
```

### Exercise:

- Download `petrol_prices.csv` and convert it into pandas dataframe.
- Plot `Year` data in the `x-axis` and `USA` data in the `y-axis`.
- You can make additional formatting (`title`, `x-label`, `y-label`, etc)

```python
import pandas as pd

df = pd.read_csv('petrol_prices.csv')

for i in df.columns[1:]:
    plt.plot(df['Year'],df[i], label=i)
plt.legend()
```

**Add `Canada` data**  
**Add marker**  
**Add labels**  

```python
import pandas as pd
df = pd.read_csv('petrol_prices.csv')

plt.plot(df['Year'], df['USA'], marker="o",label="USA")
plt.plot(df['Year'], df['Canada'], marker="*", label="Canada")
plt.plot(df['Year'], df['Australia'],label="Australia")

plt.xlabel('Year')
plt.ylabel('Petrol Data')
plt.title('Data over the Years', {'fontsize':15, 'color': 'red'})
plt.legend()
```
